# Storytelling Engine Plain-Text Test

Tests `POST /ai/storytelling_engine/game-concept` and expects a **single detailed string** response.

```python
from urllib.request import Request, urlopen
from urllib.error import HTTPError, URLError

BASE_URL = "http://localhost:3000"
USER_ID = "69b570a08e33a3fb8d0eb888"

assert USER_ID != "PUT_A_REAL_USER_ID_HERE", "Set USER_ID before running."
```

```python
import json

def call_storytelling_engine(user_id: str, base_url: str = BASE_URL):
    url = f"{base_url}/ai/storytelling_engine/game-concept"
    payload = json.dumps({"userId": user_id}).encode("utf-8")

    req = Request(
        url,
        data=payload,
        headers={"Content-Type": "application/json"},
        method="POST",
    )

    try:
        with urlopen(req, timeout=120) as response:
            text = response.read().decode("utf-8")
            return {
                "ok": True,
                "status": response.status,
                "text": text,
                "content_type": response.headers.get("Content-Type", ""),
            }
    except HTTPError as e:
        raw = e.read().decode("utf-8")
        return {
            "ok": False,
            "status": e.code,
            "error": raw,
        }
    except URLError as e:
        return {
            "ok": False,
            "status": None,
            "error": str(e),
        }
```

```python
result = call_storytelling_engine(USER_ID)
print("Status:", result["status"])

if not result["ok"]:
    print("Error:")
    print(result["error"])
else:
    print("Content-Type:", result["content_type"])
    print("Length:", len(result["text"]))
    print()
    print(result["text"])
```
